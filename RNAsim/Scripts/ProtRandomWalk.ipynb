```python
import networkx as nx

def get_hp_seq(seq: str) -> str:
  H = "VILMFWYAC"
  P = "EDRKGPSTHQN"
  hpseq = list()
  for char in seq:
    if char in H:
      hpseq.append("H")
    if char in P:
      hpseq.append("P")
    if not (char in P or char in H):
      raise RuntimeError("Not a valid aminoacid sequence.")
  return "".join(hpseq)


class AcidChain(nx.Graph):

  def __init__(self, seq, potential, **kws):
    super().__init__(**kws)
    self.seq = seq
    self.hp = get_hp_seq(seq)
    self.potential = potential
    self.step = kws['step'] if 'step' in kws else 1
    self.r = kws['size'] if 'size' in kws else (self.step / 2)
    for i in range(len(seq)):
      attr = dict(
        acid=seq[i],
        position=[0,0,self.step*i],
        size=self.r
      )
      self.add_node(i, **attr)
      if i != 0:
        self.add_edge(list(self.nodes)[i-1], list(self.nodes)[i])

  def __str__(self, *args, **kwargs):
    out = super().__str__(*args, **kwargs) + '\nSequence: '
    node_attr = nx.get_node_attributes(self, "acid")
    out += self.seq
    out += '\nHP Seq: '
    out += self.hp
    return out

  def get_neighbours(self, node, number=1):
    neighbours = [list(self.nodes)[item] \
                  for item in list(range(node - number, node + number + 1)) \
                  if item >= 0 and item < len(seq) and item != node]
    return neighbours

chain = AcidChain("MAGIIKKQILKHLSRFTKNLSPDKINLSTLKGEGELKN", None)
print(chain)
```
